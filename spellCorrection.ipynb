# 2. Creating insert, delete, substitution and transposition tables for alphabets a-z

```python
misspellings = open("misspellings.txt", encoding='utf-8')
misspellings = misspellings.readlines()[1:]
```

```python
misspellings = [misspellings[i].split(",") for i in range(len(misspellings))]

for i in range(len(misspellings)):
    misspellings[i].extend(misspellings[i][1].split("\t"))
    misspellings[i].pop(1)
    misspellings[i][1] = misspellings[i][1].strip()
    misspellings[i][-1] = misspellings[i][-1].removesuffix('\n')
```

```python
alphabets = list(string.ascii_lowercase)
alphabets.append('@')
```

```python
total_del = 0
total_ins = 0
total_sub = 0
total_tran = 0
```

```python
def insert_check(s1,s2):
    if (len(s1) > len(s2)) and (not(list(set(s2) - set(s1)))):        
        return True
    else:
        False
```

```python
def insert_t(s1,s2):
     if (len(s1) > len(s2)) and (not(list(set(s2) - set(s1)))):        
        char = list((Counter(s1)-Counter(s2)).elements())[0]
        idx = s1.index(char)
        if idx == 0:
            X = '@'
            Y = char
        else:
            X = s1[idx-1]
            Y = char
        return X,Y
```

```python
def sub_check(s1,s2):
    if (len(s1) == len(s2)) and (set(s1) != set(s2)):
        return True
    else:
        False
```

```python
def sub_t(s1,s2):
    if (len(s1) == len(s2)) and (set(s1) != set(s2)):
        X = list((Counter(s1)-Counter(s2)).elements())[0]
        Y = list((Counter(s2)-Counter(s1)).elements())[0]
        
        return X,Y
```

```python
sub = {
    alphabets[n]: {alphabets[n]: 0 for n in range(0, 26)} for n in range(0, 26)
}

for j in range(len(misspellings)):
    corr = misspellings[j][0]
    typo = misspellings[j][1:]

    for i in range(len(typo)):
        if (len(corr) == len(typo[i])) and (set(corr) != set(typo[i])):
            total_sub += 1
            
            X = list((Counter(corr)-Counter(typo[i])).elements())[0]
            Y = list((Counter(typo[i])-Counter(corr)).elements())[0]

            sub[X][Y] += 1                

df = pd.DataFrame.from_dict(sub, orient='index')
df                
```

```python
def trans_check(s1,s2):
    if (len(s1) == len(s2)) and (set(s1) == set(s2)): 
        return True
    else:
        False
```

```python
def trans_t(s1,s2):
    if (len(s1) == len(s2)) and (set(s1) == set(s2)):      
        l = list(zip(s1,s2))
        filtered = filter(lambda l: l[0] != l[1], l)
        f = list(list(filtered)[0])            
        X = f[0]
        Y = f[1]
        return X,Y
```

```python
trans = {
    alphabets[n]: {alphabets[n]: 0 for n in range(0, 26)} for n in range(0, 26)
}

for j in range(len(misspellings)):
    corr = misspellings[j][0]
    typo = misspellings[j][1:]

    for i in range(len(typo)):
        if (len(corr) == len(typo[0])) and (set(corr) == set(typo[0])):  
            total_tran += 1
            l = list(zip(corr,typo[0]))
            filtered = filter(lambda l: l[0] != l[1], l)
            f = list(list(filtered)[0])            
            X = f[0]
            Y = f[1]
            
            trans[X][Y] += 1    

df = pd.DataFrame.from_dict(trans, orient='index')
df                
```

# 3. Calculating P(x|w) using the Error Model Tables

```python
def P(X,w):
    if(del_check(X,w)):
        a,b = del_t(X,w)
        freq = delete[a][b]
        p = freq/total_del
        
    elif(insert_check(X,w)):
        a,b = insert_t(X,w)
        freq = insert[a][b]
        p = freq/total_ins
        
    elif(sub_check(X,w)):
        a,b = sub_t(X,w)
        freq = sub[a][b]
        p = freq/total_sub
        
    elif(trans_check(X,w)):
        a,b = trans_t(X,w)
        freq = trans[a][b]
        p = freq/total_tran
        
    return p
```
